> # Event
- C – Closest approach to a coast, not followed by a landfall
- G – Genesis
- I – An intensity peak in terms of both pressure and wind
- L – Landfall (center of system crossing a coastline)
- P – Minimum in central pressure
- R – Provides additional detail on the intensity of the cyclone when rapid changes are underway
- S – Change of status of the system
- T – Provides additional detail on the track (position) of the cyclone
- W – Maximum sustained wind speed


> # Status
- TD – Tropical cyclone of tropical depression intensity (< 34 knots)
- TS – Tropical cyclone of tropical storm intensity (34-63 knots)
- HU – Tropical cyclone of hurricane intensity (> 64 knots)
- EX – Extratropical cyclone (of any intensity)
- SD – Subtropical cyclone of subtropical depression intensity (< 34 knots)
- SS – Subtropical cyclone of subtropical storm intensity (> 34 knots)
- LO – A low that is neither a tropical cyclone, a subtropical cyclone, nor an extratropical cyclone (of any intensity)
- WV – Tropical Wave (of any intensity)
- DB – Disturbance (of any intensity)

```python
pacific = pd.read_csv('pacific.csv')
```

```python
pacific[['Status']]=pacific[['Status']].apply(lambda x : x.str.strip())
```

```python
pacific['Date'] = pd.to_datetime(pacific['Date'], format= '%Y%m%d')
```

```python
pacific.info()
```

gw belom ngerti re.match

```python
#  ganti latitude longitude ke float
def alp_floats(s):
    m = re.match('[+-]?\d+(\.\d+)?', s)
    return float(m.group(0)) if m is not None and m.start() == 0 else 0
Latitude_new= []
Longitude_new = []
for i in pacific['Latitude']:
    Latitude_new.append(i)
for i in pacific['Longitude']:
    Longitude_new.append(i)
pacific['Latitude'] = [alp_floats(i) for i in Latitude_new]
pacific['Longitude'] = [alp_floats(i) for i in Longitude_new]
```

```python
pacific.info()
```

```python
# Imputation
for column in pacific.columns:
    missing_cnt = pacific[column][pacific[column] == -999].count()
    print('Missing Values in column {col} = '.format(col = column) , missing_cnt )
    if missing_cnt!= 0:
#         print('in ' , column)
        mean = round(pacific[column][pacific[column] != -999 ].mean())
#         print("mean",mean)
        index = pacific.loc[pacific[column] == -999 , column].index
#         print("index" , index )
        pacific.loc[pacific[column] == -999 , column] = mean
#         print(df.loc[index , column])
```

```python
pacific
```

```python
#  ganti format jam
pacific['Time'] = pacific['Time'].astype('object')
def hhmm(time):
    time = str(time)
    digits = re.findall(r'\d', time)
    t = ''
    if len(digits) == 1:
        t ='0{i}00'.format(i =time)
    elif len(digits) == 2:
        t = '{i}00'.format(i =time)
    elif len(digits) == 3:
        t = '0{i}'.format(i =time)
    else:
        t = time
    return t
pacific['Time'] = pacific['Time'].apply(hhmm)
pacific['Time'] = pd.to_datetime(pacific['Time'] , format='%H%M').dt.time
```

```python
# ganti status jadi kategorikal
pacific['Status'] = pacific['Status'].astype('category')
```

```python
# drop data ID dan Event
pacific = pacific.drop(columns = ['ID' , 'Event'])
```
